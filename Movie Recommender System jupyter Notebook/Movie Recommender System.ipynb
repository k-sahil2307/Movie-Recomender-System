Data Preprocessing

```python
#Merge these 2 data Frames
movies = movies.merge(credits,on='title')
```

```python
movies.head(1)
```

```python
#Now I will take some columns and drop the rest
#Key what are taking

#Genres
#Id
#Keywords
#Title
#Overview
#Cast
#Crew

movies = movies[['movie_id','title','overview','genres','keywords','cast','crew']]
#This is subjective as you can chose columns according to relavance
```

```python
movies.head()
```

```python
#To check for missing data
movies.isnull().sum()
```

```python
movies.dropna(inplace = True)
```

```python
#To Check for duplicate values 
```

```python
movies.duplicated().sum()
```

```python
movies.iloc[0].genres
```

```python
#'[{"id": 28, "name": "Action"}, {"id": 12, "name": "Adventure"}, {"id": 14, "name": "Fantasy"}, {"id": 878, "name": "Science Fiction"}]'
#['Action','Adventure'.....] to be converted into this form

import ast  #it converts string of list to List

def convert(text):
    L = []
    for i in ast.literal_eval(text):
        L.append(i['name']) 
    return L 
```

```python
movies['genres'] = movies['genres'].apply(convert)
movies.head()
```

```python
movies['keywords'] = movies['keywords'].apply(convert)
movies.head()
```

```python
#in cast column let's take only first three charachters as they will be only relavant to our recomender system
def convert3(obj):
    L = []
    counter = 0
    for i in ast.literal_eval(obj):
        if counter != 3:
            L.append(i['name'])
            counter+=1
        else:
            break
    return L 
```

```python
movies['cast'] = movies['cast'].apply(convert3)
movies.head()
```

```python
#Crew : i will take only director's name from the crew
def fetch_director(text):
    L = []
    for i in ast.literal_eval(text):
        if i['job'] == 'Director':
            L.append(i['name'])
    return L 
```

```python
movies['crew'] = movies['crew'].apply(fetch_director)
```

```python
movies.head(5)
```

```python
#converting overview column from string to list
movies['overview'] = movies['overview'].apply(lambda x:x.split())
```

```python
#problem : join first and last name to make 1 tag because people can have smae first name
movies['genres'] = movies['genres'].apply(lambda x :[i.replace(" ","") for i in x])
movies['keywords'] = movies['keywords'].apply(lambda x :[i.replace(" ","") for i in x])
movies['cast'] = movies['cast'].apply(lambda x :[i.replace(" ","") for i in x])
movies['crew'] = movies['crew'].apply(lambda x :[i.replace(" ","") for i in x])
```

```python
movies.head()
```

```python
#now make a new column now as tags 
movies['tags'] = movies['overview'] + movies['genres'] + movies['keywords'] + movies['cast'] + movies['crew']
```

```python
new_df = movies[['movie_id','title','tags']]
```

```python
new_df
```

```python
#Convert tags from list to string 
new_df['tags'] = new_df['tags'].apply(lambda x :" ".join(x))
```

```python
new_df.head()
```

```python
new_df['tags'] = new_df['tags'].apply(lambda x :x.lower())
```

```python
new_df.head()
```

VECTORIZATION OF DATA 

```python
#Problem : dance,danced,dancing are same but vectors is taking it as different 
#Steming Technique  
import nltk #install this library :NLP
```

```python
from nltk.stem.porter import PorterStemmer 
ps = PorterStemmer()
```

```python
def stem(text):
    y = []
    for i in text.split():
        y.append(ps.stem(i))
    return " ".join(y)
```

```python
new_df['tags'] = new_df['tags'].apply(stem)
```

```python
#Text to Vector--Textvectorization(technique- Bag of Words)
#Bag of Words -- join all 5000 tags,count most common words after droping stop words(is,the...etc),
#                 count the occurance of most common words in all tags to maKE vector 
# This can be done by Sckit learn libraray

from sklearn.feature_extraction.text import CountVectorizer
cv = CountVectorizer(max_features=5000,stop_words='english')
```

```python
vectors = cv.fit_transform(new_df['tags']).toarray()
```
